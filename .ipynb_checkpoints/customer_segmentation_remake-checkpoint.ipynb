## Preparing Data for clustering !  

#### Get informations about products ! 

```python
df["Description"].unique()[:50]
```

```python
#get all products 
#count frequencies 

def extract_products_count(n_largest=180): 
    ''' 
    return products_list : the list of all products , count_words : a dict of 'word' : 'frequency'   
    n_largest : top n_largest most frequent words
    '''
    count_words = pd.Series()
    products_list = df["Description"].unique()
    #to remove colors from count_words
    colors = ["PINK", "RED", "BLUE", "GREEN", "WHITE"]
    
    for product in products_list: 
        s = product.split(" ") 
        #iterating trhough words & cleaning 
        for word in s: 
            if word.isnumeric() or len(word) < 3 or word in colors: 
                break
            if word in count_words: 
                count_words[word] += 1 
            else: 
                count_words[word] = 0 
                
    #returning the top 200 frequent words
    count_words = count_words.sort_values(ascending=False)[:n_largest]
    count_words = count_words.to_dict()
    
    return products_list, count_words
```

```python
#extracting products infos ...

products_list, count_words = extract_products_count()
#removing the word 'SET' because it doesn't describe the product 
del count_words["SET"]
print(f"Number of product detected : {len(products_list)}") 
print()
```

### Plot frequencies 

```python
plt.figure(figsize=(12,25))
sns.barplot(x=list(count_words.values()), y=list(count_words.keys()),orient="h")
plt.title("Words and their number of occuence")
plt.show()
```

## One hot encoding Matrix ! Preparing for clustering 

```python
X = pd.DataFrame(columns=list(count_words.keys()), index=products_list, dtype=int) 
X.fillna(0, inplace=True)
X
```

```python
def to_one_hot_vector(productName):
    ''' fill X with a row of 1 if productName contains the column title or 0 '''
    s = productName.split(" ") 
    for word in s: 
        X.loc[productName][word] = 1
```

```python
for product in products_list: 
    to_one_hot_vector(product)
```

### indicating UnitPrice Range

```python
ranges = [0,1,2,3,5,10] 
#generate columns of ranges
for i in range(len(ranges)): 
    if i == len(ranges) - 1:    
        col_name = f">=10"
    else: 
        col_name = f"[{ranges[i]},{ranges[i+1]}]"
    X[col_name] = np.zeros((len(X), 1))
```

```python
#TODO : 
#Make it faster ! 

data = df[["Description", "UnitPrice", "InvoiceDate"]]
for product in products_list: 
    #get price 
    price = data[data["Description"] == product]["UnitPrice"].mean()
    j = 0 
    while price >= ranges[j]: 
        j += 1 
        if j == len(ranges): 
            break
    if j == len(ranges): 
        #more than 10 euros 
        X.loc[product][">=10"] = 1 
    else: 
        col = f"[{ranges[j-1]},{ranges[j]}]"
        X.loc[product][col] = 1
```

```python
X_mat = X.values #or .as_matrix()
```

## Clustering time : We run Kmean clustering 

```python
#imports
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples
import matplotlib.cm as cm
from wordcloud import WordCloud
from sklearn.decomposition import PCA
```

```python
#preparing kmean clsutering 

cluster_range = range(2,10) 

for n_cluster in cluster_range: 
    kmeans = KMeans(n_clusters=n_cluster, init='k-means++', n_init=40)
    kmeans.fit(X_mat)
    clusters = kmeans.predict(X_mat)
    sil_score = silhouette_score(X_mat, clusters)
    print(f"Silhouette score for {n_cluster} clusters is :  {sil_score}")
    print("-"*50) 
    print()
```

```python
best_clusters_nb = [5,6,7,9]
cluster_label_report = {}

for n_cluster in best_clusters_nb: 
    f = plt.figure(figsize=(10,8)) 
        
    kmeans = KMeans(n_clusters=n_cluster, init='k-means++', n_init=20)
    kmeans.fit(X_mat)
    cluster_labels = kmeans.predict(X_mat) 
    cluster_label_report[n_cluster] = cluster_labels
    
    avg_sil = silhouette_score(X_mat, cluster_labels)
    print(f"avg_sil for {n_cluster} : {avg_sil}")
    sample_sil = silhouette_samples(X_mat, cluster_labels)
    
    plt.title(f"Silhouette Analysis for {n_cluster} clusters")
    plt.xlabel("Silhouette coefficient") 
    plt.ylabel("Cluster Label")

    
    y_lower = 10
    for i in range(n_cluster): 
        #get all samples in cluster i 
        i_th_cluster_values = sample_sil[cluster_labels == i ]
        i_th_cluster_values.sort() 
        size_i_th_cluster = i_th_cluster_values.shape[0]
        y_upper = y_lower + size_i_th_cluster
        color = cm.nipy_spectral(float(i) / n_cluster)
        
        plt.fill_betweenx(np.arange(y_lower, y_upper),
                          0, i_th_cluster_values,
                          facecolor=color, edgecolor=color, alpha=0.7)
        plt.text(-0.05, y_lower + 0.5 * size_i_th_cluster, str(i))
        
        y_lower = y_upper + 10
        
        plt.axvline(x=avg_sil, color="red", linestyle="--")
        plt.yticks([])  # Clear the yaxis labels / ticks
        plt.xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    
```

### Conclusion : We will drop n_cluster = 9 and we will use pca to visualize these clusters

```python
#let's keep 5 clusters
n_cluster = 5
cluster_labels = cluster_label_report[n_cluster]

X["category"] = -1
for i in range(n_cluster):
    X["category"].iloc[cluster_labels == i] = i
    size = len(X[X["category"] == i])
    print(f"{size} samples are part of category number {i} ! ") 
    print() 
```

### Visualizing these clusters with WordCloud

```python
def count_word_by_clusters(): 
    '''return a list of word,occurency dict for each cluster '''
    l = [] 
    remove = ["PINK", "RED", "BLUE", "SET", "WHITE", "GREEN"] 
    for i in range(n_cluster): 
        d = dict()
        #get all products 
        products = X[X["category"] == i].index
        for product in products: 
            s = product.split(" ") 
            for word in s: 
                if (word in remove) or (word.isnumeric()) or (len(word) < 3): 
                    break
                if word not in d: 
                    d[word] = 0
                else: 
                    d[word] += 1 
        l.append(d)
    return l
```

```python
color = [0, 160, 130, 95, 280, 40, 330, 110, 25]

def random_color_func(word=None, font_size=None, position=None,
                      orientation=None, font_path=None, random_state=None):
    h = int(360.0 * tone / 255.0)
    s = int(100.0 * 255.0 / 255.0)
    l = int(100.0 * float(random_state.randint(70, 120)) / 255.0)
    return "hsl({}, {}%, {}%)".format(h, s, l)


word_count_cluster = count_word_by_clusters()

f = plt.figure(1, figsize=(14,14))

for i in range(n_cluster): 
    tone = color[i]
    wc = WordCloud(color_func=random_color_func,background_color='white')
    wc.generate_from_frequencies(word_count_cluster[i])
    ax = f.add_subplot(4,2, i+1)
    ax.imshow(wc, interpolation="bilinear")

plt.show()
```
